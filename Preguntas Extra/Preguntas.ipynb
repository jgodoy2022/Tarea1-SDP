
1. Mergesort clásico 

El algoritmo mergesort clásico trabaja de manera recursiva, el arreglo se divide en dos mitades de tamaño similar y luego se ordena cada mitad de manera recursiva antes de fusionarlas mediante una operación de merge.

Todos los elementos del arreglo son procesados exactamente una vez durante la fase de merge. Es decir, aunque el arreglo esté dividido en varios subarreglos, la suma del trabajo de merge en ese nivel es proporcional a $n$, ya que cada elemento es comparado y copiado una sola vez dando una complejidad de:

$$O(n)$$

Ademas, el árbol de recursión tiene altura $log₂(n)$, ya que en cada nivel el tamaño del problema se reduce a la mitad, dando una complejidad de:

$$O(log(n))$$

Por lo que la complejidad total del algoritmo es:

$$O(n \log n)$$

2. K-way mergesort con merge lineal

En esta variante, el algoritmo trabaja de manera recursiva dividiendo el arreglo en $k$ subarreglos de tamaño similar en cada nivel. Luego, cada subarreglo se ordena recursivamente antes de fusionarlos mediante un merge k-way.

Durante la fase de merge, para cada uno de los $n$ elementos del arreglo se busca el mínimo entre los $k$ subarreglos activos. Esto implica que por cada elemento se realiza una comparación entre $k$ candidatos, por lo que el trabajo total en cada nivel es proporcional a $n \cdot k$, dando una complejidad de:

$$O(nk)$$

Además, el árbol de recursión tiene altura $log_k(n)$, ya que en cada nivel el tamaño del problema se reduce en un factor de $k$, dando una complejidad de:

$$O(log_k(n))$$

Por lo que la complejidad total del algoritmo es:

$$O(nk \log_k(n))$$
